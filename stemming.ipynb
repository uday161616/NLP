```python
import nltk
```

```python
words = [
    "connect", "connected", "connecting", "connection", "connections",
    "compute", "computer", "computers", "computing", "computation",
    "analyze", "analyzing", "analyzed", "analysis", "analyst", "analytics"
]
```

Porter Stemming

```python
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for x in words:
    print(stemming.stem(x))
```

Regex stemming

```python
#Regex stemming

from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$|ed$|ion$|ions$', min=4)
```

```python
for x in words:
    print(reg_stemmer.stem(x))
```

```python
#Snowball stemming
from nltk.stem import SnowballStemmer
```

```python
snow_stemmer = SnowballStemmer('english')
```

```python
for word in words:
    print(word + "--->"+snow_stemmer.stem(word))
```
