# Stemming — Definition

**Stemming** is a text normalization technique in Natural Language Processing (NLP) that reduces words to a common **base form (stem)** by removing or transforming prefixes/suffixes.
The resulting stem may **not** be a valid dictionary word (e.g., *"studies" → "studi"*), but it helps group related word variants for tasks like search, indexing, and text classification.
